```python
import pandas as pd 
from blendsql import BlendSQL
```

```python
bsql = BlendSQL(
    {
        "People": pd.DataFrame(
            {
                "Name": [],
                "Known_For": [],
            }
        ),
        "Eras": pd.DataFrame({"Years": []}),
    }
)
```

```python
bsql.visualize(
    """
    SELECT * FROM People 
    WHERE Name = {{
        LLMQA('Who wrote the song Espresso?')
    }} AND {{LLMMap('Is she a singer?')}} = TRUE
    """
)
```

```python
bsql.visualize(
    """
    SELECT * FROM People P
    WHERE P.Name IN {{
        LLMQA('First 3 presidents of the U.S?', quantifier='{3}')
    }}
    """
)
```
